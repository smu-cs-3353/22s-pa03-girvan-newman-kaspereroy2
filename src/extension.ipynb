```python
import networkx as nx
from community import community_louvain
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
g = nx.read_graphml('../data/football.graphml')
communities = community_louvain.best_partition(g) # This will compute the best partition
```

```python
pos = nx.spring_layout(g) 
cmap = cm.get_cmap('Paired', max(communities.values()) + 1)
plt.figure(3, figsize=(25,25))
nx.draw_networkx_nodes(g, pos, communities.keys(), node_size=700, cmap=cmap, node_color=list(communities.values()))
nx.draw_networkx_edges(g, pos, alpha=0.5)
nx.draw_networkx_labels(g, pos)
plt.savefig("louvain.png", dpi=80, format="png")
# plt.show
```
